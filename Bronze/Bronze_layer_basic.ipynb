# Read From CSV and Write Bronze Table

## load "cust_info.csv"

```python
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Volumes/workspace/bronze/source_systems/source_crm/cust_info.csv")

df.write.mode("overwrite").saveAsTable("workspace.bronze.crm_cust_info")
```
